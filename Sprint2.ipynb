```python
if First_element == "Horacio":
  First_value = data_base.Horacio
elif First_element == "Carlos":
  First_value = data_base.Carlos
elif First_element == "Andrea":
  First_value = data_base.Andrea
elif First_element == "Valentina":
  First_value = data_base.Valentina
elif First_element == "Jorge":
  First_value = data_base.Jorge
elif First_element == "Ivan":
  First_value = data_base.Ivan
elif First_element == "Ana":
  First_value = data_base.Ana
elif First_element == "Santiago":
  First_value = data_base.Santiago
elif First_element == "Lorena":
  First_value = data_base.Lorena
elif First_element == "Gabriel":
  First_value = data_base.Gabriel
else:
  print("Valor no disponible")

if Second_element == "Horacio":
  Second_value = data_base.Horacio
elif Second_element == "Carlos":
  Second_value = data_base.Carlos
elif Second_element == "Andrea":
  Second_value = data_base.Andrea
elif Second_element == "Valentina":
  Second_value = data_base.Valentina
elif Second_element == "Jorge":
  Second_value = data_base.Jorge
elif Second_element == "Ivan":
  Second_value = data_base.Ivan
elif Second_element == "Ana":
  Second_value = data_base.Ana
elif Second_element == "Santiago":
  Second_value = data_base.Santiago
elif Second_element == "Lorena":
  Second_value = data_base.Lorena
elif Second_element == "Gabriel":
  Second_value = data_base.Gabriel
else:
  print("Valor no disponible")
```

```python
print(First_value)
print(Second_value)
```

```python
AB = (First_value[0] * Second_value[0]) + (First_value[1] * Second_value[1]) + (First_value[2] * Second_value[2]) + (First_value[3] * Second_value[3]) + (First_value[4] * Second_value[4]) + (First_value[5] * Second_value[5])

print(AB)

Ap = math.sqrt((First_value[0]**2) + (First_value[1]**2) + (First_value[2]**2) + (First_value[3]**2) + (First_value[4]**2) + (First_value[5]**2))
Bp = math.sqrt((Second_value[0]**2) + (Second_value[1]**2) + (Second_value[2]**2) + (Second_value[3]**2) + (Second_value[4]**2) + (Second_value[5]**2))
print(Ap)
print(Bp)

Cos = AB/(Ap * Bp)
print("------------------------------------------------------------")
print("El valor del coseno es:")
print(Cos)
```
